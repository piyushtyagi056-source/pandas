# Adding Columns

```python
import pandas as pd 

data = {
    "name" : ["ram","shyam","ghanshyam","dhanshyam","aditi","jagdish","raj","simran"],
    "age" :[28,34,22,30,29,40,25,32],
    "salary" :[50000,60000,45000,52000,40000,70000,40000,50000],
    "performance_score" :[85,90,78,92,88,95,80,89]
}

df = pd.DataFrame(data)
print(df)
df['bonus'] = df['salary']*0.1  #  simple method add column at the last 
print(df)
# using the insert 
df.insert(0,'employee_id',[10,20,30,40,50,60,70,80])
df
```

# Handling  Missing Data

# Sorting and Aggeration 

```python
#df.sort_values(by='column_name,ascending = True/False,inplace = True)

import pandas as pd 

data = {
    'name' : ['arun','varun','karun'],
    'age' :[28,20,32],
    'salary' :[10000,20000,30000]
}

df = pd.DataFrame(data)

df.sort_values(by='age',ascending=False,inplace = True)
df
```

```python
#df.sort_values(by=['column_name','column 2'] ,ascending = [True/False,true/false],inplace = True)

import pandas as pd 

data = {
    'name' : ['arun','varun','karun'],
    'age' :[28,20,32],
    'salary' :[10000,20000,30000]
}

df = pd.DataFrame(data)

df.sort_values(by=['age','salary'],ascending=[True,False],inplace = True)
df
```

```python
'''
df[clmn name],mean()
df[clmn name],sum()
df[clmn name],min()
df[clmn name],max()

'''
import pandas as pd 

data = {
    'name' : ['arun','varun','karun'],
    'age' :[28,20,32],
    'salary' :[10000,20000,30000]
}

df = pd.DataFrame(data)

avg_salary = df['salary'].mean()
avg_salary
```

```python
#grouping 
#  common aggrigation fx  min max sum  std count mean last mai () jarur

import pandas as pd 

data = {
    'name' : ['arun','varun','karun','narun','marun'],
    'age' :[28,34,22,34,28],
    'salary' :[50000,60000,45000,52000,480000]
}

df = pd.DataFrame(data)

grouped = df.groupby('age')['salary'].sum()  
grouped
```

```python
# multiple column grouping 

import pandas as pd 

data = {
    'name' : ['arun','varun','karun','narun','marun'],
    'age' :[28,34,22,34,28],
    'salary' :[50000,60000,45000,52000,480000]
}

df = pd.DataFrame(data)

grouped = df.groupby(['age','name'])['salary'].sum()
grouped
```
